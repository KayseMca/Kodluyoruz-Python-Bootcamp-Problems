## Fibonnaci

```python
def fib(n):
    a,b=1,1
    toplam = 0
    fib_arr = [1]
    for i in range(n-1):
        fib_arr.append(b)
        toplam = a+b
        a = b
        b = toplam
    return fib_arr
```
